## **Recognising Handwritten Digits on MINIST Dataset using KNN**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### **Step 1. Data Preparation**

```python
df = pd.read_csv('train.csv')
print(df.shape)
```

```python
print(df.columns)
```

```python
df.head()
```

```python
data = df.values
print(data.shape)
print(type(data))
```

```python
X = data[:,1:]
Y = data[:,0] #labels
print(X.shape,Y.shape)
```

```python
# dividng data into traing and test
split = int(0.8*X.shape[0])
print(split)

X_train = X[:split,:]
Y_train = Y[:split]

X_test = X[split:,:]
Y_test = Y[split:]

print(X_train.shape,Y_train.shape)
print(X_test.shape,Y_test.shape)
```

```python
# Visualise Some Samples

def drawImg(sample):
    img = sample.reshape(28,-1)
    plt.imshow(img,cmap='gray')
    plt.show()
    
drawImg(X_train[3])    
print(Y_train[3])
```

### **Step-2 KNN**

```python
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))

def knn(X,Y,queryPoint,k=5): 
    vals = []
    m = X.shape[0]
    for i in range(m):
        d = dist(queryPoint,X[i])
        vals.append([d,Y[i]])      
    vals = sorted(vals)

    vals = vals[:k]
    vals = np.array(vals)
#     print(vals)
    new_vals = np.unique(vals[:,1],return_counts=True)
#     print("new_vals",new_vals)
    index = new_vals[1].argmax()
#     print("index",index)
    pred = new_vals[0][index]
    return pred
```

### **Step-3 Make Predictions**

```python
pred = knn(X_train,Y_train,X_test[3])
print(pred)
```

- 8400 > Q (no. of samples in testing data)<br>
<br>

So Complexity is going to be of this order O(NQ)

Training time: O(1) #bcz we are not doing not doing any thing in training.

it is a non-parametric algorithm
